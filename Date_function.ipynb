```python
from datetime import datetime

now = datetime.now()
print(now)
# Example: 2026-05-05 10:30:45.123456
```

```python
from datetime import date

today = date.today()
print(today)
# 2026-05-05
```

```python
now1 = datetime.now()

print(now1.year)   # 2026
print(now1.month)  # 5
print(now1.day)    # 5
print(now1.hour)
print(now1.minute)
print(now1.second)
```

```python
# Convert String → Date (Parsing)
date_str = "2026-05-05"

dt = datetime.strptime(date_str, "%Y-%m-%d")
print(dt)
# 2026-05-05 00:00:00
```

```python
# Convert date to string

now = datetime.now()

formatted = now.strftime("%d-%b-%Y")
print(formatted)
# 05-May-2026
```

| Format | Meaning          |
| ------ | ---------------- |
| `%Y`   | Year (2026)      |
| `%m`   | Month (05)       |
| `%d`   | Day (05)         |
| `%H`   | Hour             |
| `%M`   | Minute           |
| `%S`   | Second           |
| `%b`   | Month name (May) |


```python
# date addition
from datetime import timedelta

today = datetime.now()
future = today + timedelta(days=7)

print(future)
```

```python
# subtraction
past = today - timedelta(days=10)
past
```

```python
d1 = datetime(2026, 5, 1)
d2 = datetime(2026, 5, 5)

diff = d2 - d1
print(diff.days)
# 4
```

```python
d1 = datetime(2026, 5, 1)
d2 = datetime(2026, 5, 5)

print(d1 < d2)  # True
print(d1 == d2) # False
```

```python
dt = datetime(2026, 5, 5)

print(dt.weekday())  
# 0 = Monday, 6 = Sunday

print(dt.strftime("%A"))
# Tuesday
```

```python
now = datetime.now()

path = f"/data/year{now.year}/month{now.month}/day{now.day}"
print(path)
# /data/year2026/month5/day5
```
