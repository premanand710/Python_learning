```python
# Tutorial: Python Dictionary Basics

# 1. Creating a dictionary
my_dict = {'name': 'Alice', 'age': 25, 'city': 'New York'}

# 2. Accessing values
name = my_dict['name']

# 3. Adding a new key-value pair
my_dict['email'] = 'alice@example.com'

# 4. Updating a value
my_dict['age'] = 26

# 5. Removing a key-value pair
del my_dict['city']

# 6. Iterating over keys and values
for key, value in my_dict.items():
    print(f"{key}: {value}")

# 7. Checking if a key exists
if 'email' in my_dict:
    print("Email key exists.")

# 8. Getting all keys and values
keys = list(my_dict.keys())
values = list(my_dict.values())
print(f'keys - {keys}')
print(f'values - {values}')

# 9. Using get() to access values safely
age = my_dict.get('age', 'Not Available')
print(age)
course = my_dict.get('course', 'Not Available')
print(course)
```

```python
# List of common dictionary methods with examples

example_dict = {'a': 1, 'b': 2, 'c': 3}

# clear()
example_dict.clear()  # Removes all items
print(example_dict)  # {}

example_dict = {'a': 1, 'b': 2, 'c': 3}

# copy()
dict_copy = example_dict.copy()
print(dict_copy)  # {'a': 1, 'b': 2, 'c': 3}

# get()
value = example_dict.get('a')
print(value)  # 1

# items()
for k, v in example_dict.items():
    print(f"{k}: {v}")

# keys()
print(list(example_dict.keys()))  # ['a', 'b', 'c']

# pop()
removed = example_dict.pop('b')
print(removed)  # 2

# popitem()
example_dict = {'a': 1, 'b': 2}
item = example_dict.popitem()
print(item)  # ('b', 2) or ('a', 1)

# setdefault()
example_dict.setdefault('d', 4)
print(example_dict)  # {'a': 1, 'd': 4}

# update()
example_dict.update({'e': 5})
print(example_dict)  # {'a': 1, 'd': 4, 'e': 5}

# values()
print(list(example_dict.values()))  # [1, 4, 5]
```

```python
# Tutorial: Python Tuple Basics

# 1. Creating a tuple
my_tuple = ('apple', 'banana', 'cherry')

# 2. Accessing elements
first_item = my_tuple[0]

# 3. Slicing a tuple
slice_tuple = my_tuple[1:3]

# 4. Iterating over a tuple
for item in my_tuple:
    print(item)

# 5. Checking if an item exists
if 'banana' in my_tuple:
    print("Banana is in the tuple.")

# 6. Tuple length
length = len(my_tuple)
print(f'Tuple length: {length}')

# 7. Concatenating tuples
tuple2 = ('orange', 'grape')
combined = my_tuple + tuple2
print(combined)

# 8. Repeating tuples
repeated = my_tuple * 2
print(repeated)

# 9. Unpacking tuples
fruit1, fruit2, fruit3 = my_tuple
print(fruit1, fruit2, fruit3)

# 10. Using tuple methods
count_banana = my_tuple.count('banana')
index_cherry = my_tuple.index('cherry')
print(f'Banana count: {count_banana}')
print(f'Cherry index: {index_cherry}')
```

```python
# Tutorial: Python Set Basics

# 1. Creating a set
my_set = {1, 2, 3, 4}
print(my_set)

# 2. Adding elements
my_set.add(5)

# 3. Removing elements
my_set.remove(2)  # Raises KeyError if 2 not present
my_set.discard(6)  # Does nothing if 6 not present

# 4. Checking membership
if 3 in my_set:
    print("3 is in the set.")

# 5. Iterating over a set
for item in my_set:
    print(item)

# 6. Set length
length = len(my_set)
print(f'Set length: {length}')

# 7. Set operations
set_a = {1, 2, 3}
set_b = {3, 4, 5}
union = set_a | set_b
intersection = set_a & set_b
difference = set_a - set_b
symmetric_difference = set_a ^ set_b
print(f'Union: {union}')
print(f'Intersection: {intersection}')
print(f'Difference: {difference}')
print(f'Symmetric Difference: {symmetric_difference}')

# 8. Clearing a set
my_set.clear()
print(my_set)  # set()

# 9. Copying a set
set_copy = set_a.copy()
print(set_copy)

# 10. Set comprehension
squared = {x * x for x in range(5)}
print(squared)
```
