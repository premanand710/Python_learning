```python
# problem: Collect as many as order untill user says "DONE".

list_orders = ['Pongal','Vada']
continue_order = True

while continue_order:
    order = input('Enter your order: ')
    if order == 'DONE':
        continue_order = False
    else:
        list_orders.append(order)

print(list_orders)

for order in list_orders:
    num = input(f"how many {order} you want: ")
print("Thanks for the order")
```

# List methods
- append(element)
- extend(iterable)
- insert(index, element)
- pop([index])
- remove(element)
- clear()

```python
def print2list(list1,list2):
    print(list1)
    print(list2)
```

```python
# append

list_stud_details = [['Stud1','M',1],['Stud2','F',2],['Stud3','M',1]]
list_details = ['Prem','M','India',9.5,76,True]

print2list(list_details,list_stud_details)
```

```python
list_details.append('a')
list_stud_details.append('a')

print2list(list_details,list_stud_details)
```

```python
list_details.pop()
list_stud_details.pop()

print2list(list_details,list_stud_details)
```

```python
list_details.pop(0)
list_stud_details.pop(0)

print2list(list_details,list_stud_details)
```

```python

list_details.append('b')
list_stud_details.append('b')

list_details.append('b')
list_stud_details.append('b')
print2list(list_details,list_stud_details)

list_details.remove('b')
list_stud_details.remove('b')

print2list(list_details,list_stud_details)
```

```python
list_details.extend(['c','d'])
list_stud_details.extend(['c','d'])
print2list(list_details,list_stud_details)
```

```python
list_details.append(['c','d'])
list_stud_details.append(['c','d'])
print2list(list_details,list_stud_details)
```

```python
list_details.insert(1,['x','y'])
list_stud_details.insert(1,['x','y'])
print2list(list_details,list_stud_details)
```

```python
print(list_details)
list_details.reverse()
print(list_details)
```
