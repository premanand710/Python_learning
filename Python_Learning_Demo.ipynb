```python
list_num1 = list(range(15))
print(list_num1)
list_num1 = list(range(2,15))
print(list_num1)
list_num1 = list(range(2,15,4))
print(list_num1)


list_num1 = list(range(10,2)) # invalid
print(list_num1)

list_num1 = list(range(10,2,-1)) 
print(list_num1)

list_num1 = list(range(10,2,-2)) 
print(list_num1)
```

# Slicing

```python
list_num3 = list(range(11))
print(list_num3)

print(list_num3[0])
print(list_num3[1:])
print(list_num3[1:4])
print(list_num3[1:8:2])
print(list_num3[::])
print(list_num3[::-1])
```

```python
list_stud_details = [['Stud1','M',"1"],['Stud2','F',"2"],['Stud3','M',"1"]]
print(list_stud_details)
print(len(list_stud_details))
```

```python
print(list_stud_details[1])
print(list_stud_details[1][1:])
print(list_stud_details[1][1:][1])

print(list_stud_details, type(list_stud_details))
print(list_stud_details[0], type(list_stud_details[0]))
print(list_stud_details[0][1], type(list_stud_details[0][1]))
```
